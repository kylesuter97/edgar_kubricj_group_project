```python
# Import BeautifulSoup
import requests
from bs4 import BeautifulSoup
```

```python
url = r'https://en.wikipedia.org/wiki/S%26P_100'
r = requests.get(url)
r.text[:1000]            
```

```python
# First, we create a soup object from the html
# The soup object allows us to navigate through the html

soup = BeautifulSoup(r.text, 'html.parser') 
```

```python
# Let us find the "Basic HTML" table in this soup object using its tag and attributes:

tag = 'table'
attributes = {'id':'constituents'}           # Note that the attribute is written as a dictionary here
table_soup = soup.find(tag, attributes)                     # the .find() method searches for elements with the given details
```

```python
# Take a peek
print(table_soup)
```

```python
table_data = []
for row in table_soup.find_all('tr'):
    row_text = [e.text.strip() for e in row.find_all('td')]
    table_data.append(row_text)

# Print out the table contents
print(table_data)
```
